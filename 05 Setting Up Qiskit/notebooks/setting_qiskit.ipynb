Add your IBM token to `.env` file. <br>
This is how your `.env` file must look

```
IBMQTOKEN = <add your token here>
```

````python
import os
from dotenv import load_dotenv
from qiskit import IBMQ

load_dotenv()
````

````python
IBMQ.save_account(os.getenv("IMBQTOKEN"))
````

````python
IBMQ.load_account()
````
